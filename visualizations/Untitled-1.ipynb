```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
data =pd.read_csv ("Automobile.csv")
data
```

```python
x= data["acceleration"]
y=data["horsepower"]
```

```python
plt.plot(x,y,color='green',marker='o',linestyle='solid')
plt.title="line chart"
plt.ylabel=("accelaration")
plt.xlabel=('horsepower')
plt.show()
```
