```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report 
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('output.csv')
data
```

```python
data = data.drop(['location' , 'mileage' , 'power'  ,'seats'      ] ,axis=1)
```

```python
data.dropna( axis=1)
```

```python
data['transmission'] = data['transmission'].replace({'manual': 1, 'automatic': 0})

data['fuel'] = data['fuel'].replace({'petrol':1, 'diesel':0})
data['type'] = data['type'].replace({'hatchback':1, 'sedan':0})
data = data.drop(['engine'], axis=1)
data.tail()
```

```python
X= data.drop('price', axis=1)
y = data['price']
X
```

```python
X_train,X_test, y_train,y_test =train_test_split(X,y, train_size=.3, random_state=42)

```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply LabelEncoder to 'brand' column
label_encoder = LabelEncoder()
data['brand'] = label_encoder.fit_transform(data['brand'])
from sklearn.preprocessing import OneHotEncoder

# One-hot encode 'brand' column
one_hot = OneHotEncoder(sparse=False)
encoded = one_hot.fit_transform(data[['brand']])

# Convert the result to a DataFrame and join with original data
encoded_df = pd.DataFrame(encoded, columns=one_hot.get_feature_names_out(['brand']))
data = data.join(encoded_df).drop('brand', axis=1)


```
